```python
feature_cols = ['log_alcohol','density','fixed acidity','log_vol_acidity','citric acid','log_res_sugar','chlorides','free sulfur dioxide','total sulfur dioxide','pH','log_sulphates']
X = df [feature_cols]
Y = df ['quality']
```

```python
 X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
```

```python
regr = linear_model.LinearRegression()
```

```python
regr.fit(X_train, Y_train)
```

```python
Y_pred = regr.predict(X_test)
```

```python
#Prints y intercept
print(regr.intercept_)
```

```python
#Prints beta coefficients
print(regr.coef_)
```

```python
print(metrics.mean_absolute_error(Y_test,Y_pred))
```

```python
print(metrics.mean_squared_error(Y_test,Y_pred))
```

```python
print(np.sqrt(metrics.mean_squared_error(Y_test,Y_pred)))
```

```python
print('Variance score: %.2f' % metrics.r2_score(Y_test, Y_pred))
```

```python
import statsmodels.api as sm
model = sm.OLS(Y_train, X_train).fit()
predictions = model.predict(X)
model.summary()
```

```python
feature_cols = ['log_alcohol','density','fixed acidity','volatile acidity','log_res_sugar','free sulfur dioxide','total sulfur dioxide','log_sulphates']
X = df [feature_cols]
Y = df ['quality']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
```

```python
model = sm.OLS(Y_train, X_train).fit()
predictions = model.predict(X_test)
model.summary()
```

```python
from  matplotlib import pyplot
#pyplot.scatter(X_test['alcohol'],Y_test)
#pyplot.plot(np.sort(X_test['alcohol'],axis=0),predictions,'r')
pyplot.scatter(Y_test,predictions)
pyplot.xlabel('Actual')
pyplot.ylabel('Predicted')
```

```python
residuals = Y_test - predictions
plt.scatter(X_test['log_alcohol'],residuals)
plt.xlabel("log_alcohol")
plt.ylabel("residuals")
plt.show()
```

```python
from statsmodels.sandbox.regression.predstd import wls_prediction_std
_, confidence_interval_lower, confidence_interval_upper = wls_prediction_std(model)
fig, ax = plt.subplots(figsize=(10,7))
ax.plot(X_train['log_alcohol'], Y_train, 'o', label="data")
ax.plot(X_train['log_alcohol'], model.fittedvalues, 'g--.', label="OLS")
#ax.plot(x, confidence_interval_upper, 'r--')
ax.plot(X_train['log_alcohol'], confidence_interval_upper, 'r--')
ax.plot(X_train['log_alcohol'], confidence_interval_lower, 'r--')
ax.legend(loc='best');

```
